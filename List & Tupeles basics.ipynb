# Lists


*   List are mutable objects
*   List can be indexed in the same way as strings
*   List can contains various datatypes

```python
L1 = ['Pradeep','abc', 34.5, 3, 'xyz',True]
print(L1)


L1[0] = 'Pqr'# since list are mutable we can change the object present at index 0


print(L1)

```

**Basic Functions used in list**

*   List.sort() used to sort the list of NUMBERS only in ascending order
*   List.appened(object) append the object at the end of the list
*   List.reverse() Reverses the list
*   List.insert(position, object) Inserts the object at the specified position in the list
*   List.pop(index) Pops the obects at the index and deletes it from the list
*   List.remove(object) Searches the list and removes the object specified only from the first location that it encounters
E.G : List = ["pradeep","pradeep"]
if you use List.remove("pradeep") it removes the pradeep present at 0 index only


```python
print(L1)

L1.append(3)

print(L1)

print(L1.count("Pradeep"))

L1.remove("Pradeep")

L1.pop(3)

print(L1)
```

# Problems

```python
#Program to store fruits entered by the user to a list

n = int(input("Enter the number of fruits "))

fruits = []

for i in range(n):
  fruits.append(input(f"Enter the {i} fruit name "))

print(fruits)
```

```python
#Program to take input of marks of the student and the display those in sorted manner

n= int(input("ENter the number of students "))

marks = []

for i in range(n):
  marks.append(int(input(f"ENter the {i+1} student's marks ")))

marks.sort()

print(marks)
```
